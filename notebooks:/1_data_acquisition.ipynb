Loads the expected CSV inputs from `data/`.
Replace these CSVs with real 1-year 5-min NIFTY data for final submission.


```python

import pandas as pd
from pathlib import Path

DATA_DIR = Path("../data")

spot = pd.read_csv(DATA_DIR/"nifty_spot_5min.csv", parse_dates=["timestamp"])
fut  = pd.read_csv(DATA_DIR/"nifty_futures_5min.csv", parse_dates=["timestamp"])
opt  = pd.read_csv(DATA_DIR/"nifty_options_5min.csv", parse_dates=["timestamp"])

print("Spot:", spot.shape, "Futures:", fut.shape, "Options:", opt.shape)
spot.head()
```

## Sanity checks

```python

assert {"open","high","low","close","volume"}.issubset(set(spot.columns))
assert "open_interest" in fut.columns
assert {"call_iv","put_iv","call_oi","put_oi","call_volume","put_volume"}.issubset(set(opt.columns))
```
