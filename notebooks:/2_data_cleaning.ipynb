# 2. Data Cleaning & Alignment

Cleans and aligns timestamps across spot, futures, options.


```python

import pandas as pd
from pathlib import Path
from src.data_utils import clean_timeseries, continuous_futures

DATA_DIR = Path("../data")

spot = pd.read_csv(DATA_DIR/"nifty_spot_5min.csv", parse_dates=["timestamp"])
fut  = pd.read_csv(DATA_DIR/"nifty_futures_5min.csv", parse_dates=["timestamp"])
opt  = pd.read_csv(DATA_DIR/"nifty_options_5min.csv", parse_dates=["timestamp"])

spot = clean_timeseries(spot)
fut  = clean_timeseries(fut)
opt  = clean_timeseries(opt)

if "contract" in fut.columns:
    fut = continuous_futures(fut)

common_ts = set(spot["timestamp"]).intersection(fut["timestamp"]).intersection(opt["timestamp"])
spot = spot[spot["timestamp"].isin(common_ts)].sort_values("timestamp")
fut  = fut[fut["timestamp"].isin(common_ts)].sort_values("timestamp")
opt  = opt[opt["timestamp"].isin(common_ts)].sort_values("timestamp")

print("Aligned:", spot.shape, fut.shape, opt.shape)
```
